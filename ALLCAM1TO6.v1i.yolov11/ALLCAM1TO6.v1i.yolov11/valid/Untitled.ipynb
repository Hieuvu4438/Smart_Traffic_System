```python
import os

# Đường dẫn thư mục
labels_dir = 'labels'  # Thư mục chứa các file txt
images_dir = 'images'  # Thư mục chứa ảnh tương ứng

# Quy tắc chuyển đổi class mới
class_mapping = {
    '1': '1',    # Class 1 -> 1 (car)
    '2': '1',    # Class 2 -> 1 (car)
    '5': '1',    # Class 5 -> 1 (car)
    '3': '2',    # Class 3 -> 2 (motorbike)
    '4': '2'     # Class 4 -> 2 (motorbike)
    # Các class khác sẽ bị bỏ qua
}

def process_txt_file(txt_path):
    modified_lines = []
    
    with open(txt_path, 'r') as f:
        lines = f.readlines()
    
    for line in lines:
        parts = line.strip().split()
        if len(parts) >= 1:  # Đảm bảo có ít nhất class id
            old_class = parts[0]
            
            # Áp dụng quy tắc chuyển đổi
            if old_class in class_mapping:
                parts[0] = class_mapping[old_class]
                modified_lines.append(' '.join(parts) + '\n')
            # Các class không nằm trong mapping sẽ bị bỏ
    
    # Ghi lại file
    with open(txt_path, 'w') as f:
        f.writelines(modified_lines)

def main():
    # Kiểm tra thư mục labels
    if not os.path.exists(labels_dir):
        print(f"Thư mục {labels_dir} không tồn tại!")
        return
    
    # Xử lý từng file txt
    processed_files = 0
    for filename in os.listdir(labels_dir):
        if filename.endswith('.txt'):
            txt_path = os.path.join(labels_dir, filename)
            process_txt_file(txt_path)
            processed_files += 1
            print(f"Đã xử lý: {filename}")
    
    print(f"\nHoàn thành! Đã xử lý {processed_files} file txt.")
    print("Quy tắc chuyển đổi:")
    print("- Class 1,2,5 -> 1 (car)")
    print("- Class 3-4 -> 2 (motorbike)")
    print("- Các class khác đã bị xóa")

if __name__ == "__main__":
    main()
```
