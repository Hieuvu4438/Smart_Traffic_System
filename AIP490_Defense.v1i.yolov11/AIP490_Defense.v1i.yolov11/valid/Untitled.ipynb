```python
import os

# Đường dẫn thư mục
labels_dir = 'labels'  # Thư mục chứa các file txt

def process_txt_file(txt_path):
    """Xử lý từng file txt để xóa các nhãn class 3"""
    modified_lines = []
    
    with open(txt_path, 'r') as f:
        lines = f.readlines()
    
    for line in lines:
        parts = line.strip().split()
        if len(parts) >= 1:  # Đảm bảo có ít nhất class id
            class_id = parts[0]
            
            # Giữ lại tất cả các dòng trừ class 3
            if class_id != '3':
                modified_lines.append(line)
    
    # Ghi lại file
    with open(txt_path, 'w') as f:
        f.writelines(modified_lines)

def main():
    # Kiểm tra thư mục labels
    if not os.path.exists(labels_dir):
        print(f"Thư mục {labels_dir} không tồn tại!")
        return
    
    # Xử lý từng file txt
    processed_files = 0
    deleted_count = 0
    
    for filename in os.listdir(labels_dir):
        if filename.endswith('.txt'):
            txt_path = os.path.join(labels_dir, filename)
            
            # Đếm số dòng trước khi xóa
            with open(txt_path, 'r') as f:
                before = len(f.readlines())
            
            process_txt_file(txt_path)
            
            # Đếm số dòng sau khi xóa
            with open(txt_path, 'r') as f:
                after = len(f.readlines())
            
            deleted = before - after
            deleted_count += deleted
            processed_files += 1
            
            print(f"Đã xử lý: {filename} (Xóa {deleted} nhãn class 3)")
    
    print(f"\nHoàn thành! Đã xử lý {processed_files} file txt.")
    print(f"Tổng số nhãn class 3 đã xóa: {deleted_count}")

if __name__ == "__main__":
    main()
```
